```python
f_cevns = ROOT.TFile.Open("../data/cohar750_cevns.root")
h_cevns = f_cevns.Get("h_cevns")
h_cevnsf90 = f_cevns.Get("h_cevnsadjf901")
```

```python
hs = ROOT.THStack("hs", "COH-Ar-750 Signal Predictions;Energy [keVee];Counts/SNS-year")
h_ar39f90.SetFillColor(ROOT.kBlue+1)
hs.Add(h_ar39f90)
hog_phase2.SetFillColor(ROOT.kGreen+1)
hs.Add(hog_phase2)
pb210.SetFillColor(ROOT.kYellow+1)
hs.Add(pb210)
concrete.SetFillColor(ROOT.kOrange+1)
hs.Add(concrete)
# neutrons.SetFillColor(ROOT.kSpring+1)
# hs.Add(neutrons)
h_testing.SetFillColor(ROOT.kSpring+1)
hs.Add(h_testing)
h_cevnsf90.SetFillColor(ROOT.kRed+1)
hs.Add(h_cevnsf90)
hlist = hs.GetHists()
bin_centers = np.arange(1, 42) - 0.5
bin_content = np.zeros(40)
for h in hlist:
    for i in range(1, 41):
        bin_content[i-1] += h.GetBinContent(i)
bin_errs = np.sqrt(bin_content)
ss_content = np.zeros(40)
for h in hlist[:4]:
    for i in range(1, 41):
        ss_content[i-1] += h.GetBinContent(i)
ss_errs = np.sqrt(ss_content)
g_err = ROOT.TGraphErrors(len(bin_content), bin_centers, bin_content, np.zeros(40), bin_errs)
rel_errs = bin_errs / bin_content
print(bin_errs)
print(bin_content)
print(rel_errs)
leg = ROOT.TLegend(0.7, 0.7, 0.9, 0.9)
leg.AddEntry(h_ar39f90, "Ar39", "f")
leg.AddEntry(hog_phase2, "HOG Phase 2", "f")
leg.AddEntry(pb210, "Pb210", "f")
leg.AddEntry(concrete, "Concrete", "f")
# leg.AddEntry(neutrons, "BRNs", "f")
leg.AddEntry(h_testing, "BRNs", "f")
leg.AddEntry(h_cevnsf90, "CEvNS", "f")
c = ROOT.TCanvas()
c.SetLogy()
hs.Draw("hist")
hs.SetMinimum(1)
leg.Draw()
g_err.Draw("P")
c.Draw()
```
